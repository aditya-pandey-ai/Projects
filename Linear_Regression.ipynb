```python
train_data = []
with open(csv_file_path1, "r") as train_file:
    next(train_file)
    for line in train_file:
        x, y = line.strip().split(",")
        train_data.append((float(x), float(y)))

test_data = []
with open(csv_file_path2, "r") as test_file:
    next(test_file)
    for line in test_file:
        x, y = line.strip().split(",")
        test_data.append((float(x), float(y)))
```

```python
def cost_function(X, Y, m, c):
    N, total_error = len(X), 0.0

    for i in range(N):
        total_error += (Y[i] - (m * X[i] + c)) ** 2

    return total_error / (2 * float(N))
```

```python
def gradient_descent(X, Y, m, c, alpha):
    dl_dm, dl_dc, N = 0.0, 0.0, len(X)

    for i in range(N):
        dl_dm += -X[i] * (Y[i] - (m * X[i] + c))
        dl_dc += -(Y[i] - (m * X[i] + c))

    m = m - (1 / float(N)) * dl_dm * alpha
    c = c - (1 / float(N)) * dl_dc * alpha

    return m, c
```

```python
def train(X, Y, m, c, alpha, n_iter):
    for i in range(n_iter):
        m, c = gradient_descent(X, Y, m, c, alpha)

        if i % 400 == 0:
            print("iteration:", i, "cost: ", cost_function(X, Y, m, c))

    return m, c
```

```python
def predict(x, m, c):
    return x * m + c
```

```python
m_train, c_train = train(X_train, Y_train, 0.0, 0.0, 0.0001, 7000)
x_new = 50.0
y_new = predict(x_new, m_train, c_train)
print(y_new)
```
