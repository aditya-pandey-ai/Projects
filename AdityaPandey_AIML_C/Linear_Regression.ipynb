```python
csv_file_path = '/content/drive/My Drive/gdsc_task/train.csv'
```

```python
data = []
with open(csv_file_path, "r") as file:
    next(file)
    for line in file:
        x, y = line.strip().split(",")
        data.append((float(x), float(y)))
```

```python
X = [item[0] for item in data]
Y = [item[1] for item in data]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, s=5)
plt.grid()
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
def cost_function(X, Y, m, c):
    N, total_error = len(X), 0.0

    for i in range(N):
        total_error += (Y[i] - (m * X[i] + c)) ** 2

    return total_error / (2 * float(N))
```

```python
def gradient_descent(X, Y, m, c, alpha):
    dl_dm, dl_dc, N = 0.0, 0.0, len(X)

    for i in range(N):
        dl_dm += -X[i] * (Y[i] - (m * X[i] + c))
        dl_dc += -(Y[i] - (m * X[i] + c))

    m = m - (1 / float(N)) * dl_dm * alpha
    c = c - (1 / float(N)) * dl_dc * alpha

    return m, c
```

```python
def train(X, Y, m, c, alpha, n_iter):
    for i in range(n_iter):
        m, c = gradient_descent(X, Y, m, c, alpha)

        if i % 400 == 0:
            print("iteration:", i, "cost: ", cost_function(X, Y, m, c))

    return m, c
```

```python
def predict(x, m, c):
    return x * m + c
```

```python
m, c = train(X, Y, 0.0, 0.0, 0.0001, 7000)
x_new = 50.0
y_new = predict(x_new, m, c)
print(y_new)
```
